```python
from sklearn.naive_bayes import *
import pandas as pd
import numpy as np

# create data frame containing your data, each column can be accessed # by df['column   name']
df = dfMaster

df['label']=0
df.label[df.DepDelay>=15]=1
df.label[df.DepDelay<15]=0
# df.ix[0:6]['label']=1
df.iloc[:,[15,29]].head()
```

```python
target_names = np.array(['Delayed','Not Delayed'])

# add columns to your data frame
df['is_train'] = np.random.uniform(0, 1, len(df)) <= 0.75

# define training and test sets
train = df[df['is_train']==True]
test = df[df['is_train']==False]

# # train_x, train_y = data[is_train], y[is_train]
# # test_x, test_y = data[is_train==False], y[is_train==False]

trainTargets = np.array(train['label']).astype(int)
testTargets = np.array(test['label']).astype(int)

# # columns you want to model
features = df.columns[0:7]

# # # call Gaussian Naive Bayesian class with default parameters
gnb = GaussianNB()

# # # train model
y_gnb = gnb.fit(train[features], trainTargets).predict(test[features])

test['pred_label']=y_gnb

# print test['label']
test.iloc[:,[15,29,31]].head()
# print y_gnb
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(test['label'], test['pred_label'])
```
